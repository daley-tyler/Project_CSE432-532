```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import soundfile as sf
import os
```

```python
#Path to data folder
data_path = "/Users/tyler/CSE432_Fork_Project/Project_CSE432-532/SER_Project/data"

speeches = []

#Loops through the Actor Folders 
for actors in range(1,25):
    actors_str = f"{actors:02d}"
    actor_Folder = os.path.join(data_path, f"Actor_{actors_str}")

    #Reads files in the different actor folders
    for filename in os.listdir(actor_Folder):
        #Makes sure its a speech and not a song
        if filename.startswith("03-01-"):
            file_path = os.path.join(actor_Folder, filename)
            speeches.append(file_path)

print("Speech Files:", len(speeches) )

songs = []

for actors in range(1,25):
    actors_str = f"{actors:02d}"
    actor_Folder = os.path.join(data_path, f"Actor_{actors_str}")

    #Reads files in the different actor folders
    for filename in os.listdir(actor_Folder):
        #Makes sure its a speech and not a song
        if filename.startswith("03-02-"):
            file_path = os.path.join(actor_Folder, filename)
            songs.append(file_path)

print("Song Files:", len(songs) )
```
