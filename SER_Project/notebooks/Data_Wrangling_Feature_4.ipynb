```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Filename number codes to labels

modality_label = {"01": "full-AV", "02": "video-only", "03": "audio-only"}
vocal_channel_label = {"01": "speech", "02": "song"}
emotion_label = {"01": "neutral", "02": "calm", "03": "happy", "04": "sad", "05": "angry", "06": "fearful", "07": "disgust", "08": "suprised"}
intensity_label = {"01": "normal", "02": "strong"}
statement_label = {"01": "Kids are talking by the door","02": "Dogs are sitting by the door"}
repetition_label = {"01": "1st", "02": "2nd"}
```

```python
# Goes through the data folder to get the wav paths

data_folder = "../data"

wav_files = []

for root, folders, files in os.walk(data_folder):
    for filename in files:
        if filename.endswith(".wav"):
            fpath = os.path.join(root, filename)
            wav_files.append(fpath)

wav_files.sort()

print("Num of wav files: ", len(wav_files))
```

```python
# Parse the filename for its parts 
def parse_filename(fpath):
    filename = os.path.basename(fpath)
    file_noWAV = filename[:-4]
    split_name = file_noWAV.split("-")

    modality_p = split_name[0]
    vocal_channel_p = split_name[1]
    emotion_p = split_name[2]
    intensity_p = split_name[3]
    statement_p = split_name[4]
    repetition_p = split_name[5]
    actor_p = split_name[6]

    actor = int(actor_p)


    if actor % 2 == 1:
        gender = "male"
    else:
        gender = "female"

    # Stores the number codes and word codes in a row for a file
    row = {"path": fpath, "filename": filename, "modality_code": modality_p, "vocal_channel_code": vocal_channel_p, "emotion_code": emotion_p, "intensity_code": intensity_p, "statement_code": statement_p, "repetition_code": repetition_p, "actor_code": actor_p, "modality": modality_label[modality_p], "vocal_channel": vocal_channel_label[vocal_channel_p], "emotion": emotion_label[emotion_p], "intensity": intensity_label[intensity_p], "statement": statement_label[statement_p], "repetition": repetition_label[repetition_p], "actor": actor, "gender": gender}

    return row
```

```python
metadata_rows = []

for fpath in wav_files:
    row = parse_filename(fpath)
    metadata_rows.append(row)

metadata_columns = ["path", "filename", "modality_code", "vocal_channel_code", "emotion_code", "intensity_code", "statement_code", "repetition_code", "actor_code", "modality", "vocal_channel", "emotion", "intensity", "statement", "repetition", "actor", "gender"]

metadata_df = pd.DataFrame(metadata_rows, columns=metadata_columns)

print("Files: ", len(metadata_df))

metadata_df.head()
```
