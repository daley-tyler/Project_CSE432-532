```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import soundfile as sf
import os
```

```python
#Path to data folder
data_path = "/Users/tyler/CSE432_Fork_Project/Project_CSE432-532/SER_Project/data"

speeches = []

#Loops through the Actor Folders 
for actors in range(1,25):
    actors_str = f"{actors:02d}"
    actor_Folder = os.path.join(data_path, f"Actor_{actors_str}")

    #Reads files in the different actor folders
    for filename in os.listdir(actor_Folder):
        #Makes sure its a speech and not a song
        if filename.startswith("03-01-"):
            file_path = os.path.join(actor_Folder, filename)
            speeches.append(file_path)

print("Speech Files:", len(speeches) )

songs = []

for actors in range(1,25):
    actors_str = f"{actors:02d}"
    actor_Folder = os.path.join(data_path, f"Actor_{actors_str}")

    #Reads files in the different actor folders
    for filename in os.listdir(actor_Folder):
        #Makes sure its a speech and not a song
        if filename.startswith("03-02-"):
            file_path = os.path.join(actor_Folder, filename)
            songs.append(file_path)

print("Song Files:", len(songs) )

#Combining speech and song
all_files = speeches + songs

#Table of file/folder names and paths
actors_df = pd.DataFrame({"path": all_files})
actors_df["filename"] = actors_df["path"].str.split("/").str[-1]
actors_df["actor_folder"] = actors_df["path"].str.split("/").str[-2]

actors_df.head()

#File Parse
par = actors_df["filename"].str.extract()

par.columns = ["Modality_Codes","Channel_Codes","Emotion_Codes","Intesity_Codes","Statement_Codes","Repetition_Codes","Actor_Codes"]

#Turning Numeric into Labels
actors_df["modality"] = actors_df["Modality_Codes"].replace({"01":"full-AV", "02":"video-only","03":"audio-only"})
actors_df["vocal_channel"] = actors_df["Channel_Codes"].replace({"01":"speech", "02":"song"})
actors_df["emotion"] = actors_df["Emotion_Codes"].replace({"01":"neutral", "02":"calm","03":"happy", "04":"sad", "05":"angry","06":"fearful","07":"disgust", "08":"surprised"})
actors_df["intensity"] = actors_df["Intesity_Codes"].replace({"01":"normal", "02":"strong"})
actors_df["statement"] = actors_df["Statement_Codes"].replace({"01":"Kids are talking by the door", "02":"Dogs are sitting by the door"})
actors_df["repetition"] = actors_df["Repetition_Codes"].replace({"01":"1st", "02":"2nd"})
#actors_df["actor"] = actors_df["Actor_Codes"].replace({})

```
