```python
import os
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 

import sys
sys.path.append("..")

from minilearn import train_test_split, StandardScaler
from minilearn import LogisticRegression
from minilearn import GaussianNaiveBayes
from minilearn import KNNClassifier
from minilearn import accuracy_score

from sklearn.linear_model import LogisticRegression as SKLogisticRegression
from sklearn.naive_bayes import GaussianNB as SKGaussianNB
from sklearn.neighbors import KNeighborsClassifier as SKKNN
```

```python
fpath = os.path.join("../data", "week5_audio.csv")

feature_df = pd.read_csv(fpath)

feature_df.head()
```

```python
m_cols = ["path", "filename", "modality_code", "vocal_channel_code", "emotion_code", "intensity_code", "statement_code", "repetition_code", "actor_code", "modality", "vocal_channel", "emotion", "intensity", "statement", "repetition", "actor", "gender"]
f_cols = []

for c in feature_df.columns:
    if c not in m_cols:
        f_cols.append(c)

X = feature_df[f_cols]
y = feature_df["emotion"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y)

scaler = StandardScaler()

X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
#Minilearn LR

ml_lr = LogisticRegression(learning_rate=0.1, n_iter=500)
ml_lr.fit(X_train_scale, y_train)
ml_lr_pred = ml_lr.predict(X_test_scale)
ml_lr_as = accuracy_score(y_test, ml_lr_pred)

print("Minilearn Logistic Regression Accuracy Score", ml_lr_as)

#Minilearn GB
ml_gb = GaussianNaiveBayes()
ml_gb.fit(X_train_scale, y_train)
ml_gb_pred = ml_gb.predict(X_test_scale)
ml_gb_as = accuracy_score(y_test, ml_gb_pred)

print("Minilearn GB Accuracy Score: ", ml_gb_as)

#Minilearn KNN
ml_knn = KNNClassifier(n_neighbors=5)
ml_knn.fit(X_train_scale, y_train)
ml_knn_pred = ml_knn.predict(X_test_scale)
ml_knn_as = accuracy_score(y_test, ml_knn_pred)

print("Minilearn KNN Accuracy Score: ", ml_knn_as)
```

```python
#SKlearn LR
sk_lr = SKLogisticRegression()
sk_lr.fit(X_train_scale, y_train)
sk_lr_pred = sk_lr.predict(X_test_scale)
sk_lr_as = accuracy_score(y_test, sk_lr_pred)

print("SKlearn Logistic Regression Accuracy Score: ", sk_lr_as)

#SKlearn GB
sk_gb = SKGaussianNB()
sk_gb.fit(X_train_scale, y_train)
sk_gb_pred = sk_gb.predict(X_test_scale)
sk_gb_as = accuracy_score(y_test, sk_gb_pred)

print("SKlearn GB Accuracy Score", sk_gb_as)

#SKlearn KNN
sk_knn = SKKNN()
sk_knn.fit(X_train_scale, y_train)
sk_knn_pred = sk_knn.predict(X_test_scale)
sk_knn_as = accuracy_score(y_test, sk_knn_pred)

print("SKlearn KNN Accuracy Score: ", sk_knn_as)
```
