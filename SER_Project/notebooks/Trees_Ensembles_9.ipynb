```python
import os
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import sys
sys.path.append("..")

from minilearn import StandardScaler, train_test_split
from minilearn import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from minilearn import DecisionTreeClassifier as MiniTree

from sklearn.tree import DecisionTreeClassifier as SKTree
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
```

```python
fpath = os.path.join("../data", "week5_audio.csv")

feature_df = pd.read_csv(fpath)

feature_df.head()
```

```python
m_cols = ["path", "filename", "emotion", "intensity", "actor", "gender", "vocal_channel"]
f_cols = []

for c in feature_df.columns:
    if c not in m_cols:
        f_cols.append(c)

X = feature_df[f_cols]
y = feature_df["emotion"]

print("Features: ", X.shape[1])
```

```python
# Split and Scale
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y)

scaler = StandardScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
models_list = [["My Minilearn CART", MiniTree(max_depth=4, min_samples_split=5), "max_depth=4"], ["SKlearn DT", SKTree(max_depth=4,random_state=42), "max_depth=4"], ["SKlearn Random Forest", RandomForestClassifier(n_estimators=100, max_depth=8, random_state=123), "n_estimators=100, max_depth=8"], ["SKlearn AdaBoost", AdaBoostClassifier(estimator=SKTree(max_depth=1, random_state=123), n_estimators=50, learning_rate=0.5), "n_estimators=50, learning_rate=0.5"]]

result = []
preds = {}

for row in models_list:
    name = row[0]
    model = row[1]
    parameters = row[2]
    
    model.fit(X_train_scale, y_train)
    pred = model.predict(X_test_scale)
    pred_train = model.predict(X_train_scale)
    preds[name] = pred

    result.append({"model": name, "parameters": parameters, "train_accuracy": accuracy_score(y_train, pred_train), "test_accuracy": accuracy_score(y_test, pred), "w_precision": precision_score(y_test, pred, avg="weighted"), "w_recall": recall_score(y_test, pred, avg="weighted"), "w_F1": f1_score(y_test, pred, avg="weighted")})

result_df = pd.DataFrame(result)
result_df = result_df.sort_values("w_F1", ascending=False)

result_df
```

```python
b_model = result_df.iloc[0]["model"]
b_pred = preds[b_model]
labels = np.unique(y_test)
cmatrix = confusion_matrix(y_test, b_pred, labels=labels)

plt.figure(figsize=(10,5))
sns.heatmap(cmatrix, annot=True, xticklabels=labels, yticklabels=labels)
```
